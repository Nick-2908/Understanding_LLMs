```python
text=open('input.txt').read()
```

```python
chars = sorted(list(set(text)))
vocab_size = len(chars)
```

```python
print(chars)
```

```python
print("(" in chars)
```

```python
stoi={ch:i for i,ch in enumerate(chars)}
itos={i:ch for i,ch in enumerate(chars)}



encode=lambda s:[stoi[c] for c in s]
decode=lambda l:''.join([itos[i] for i in l])
```

```python
print(encode("hello"))
print(decode(encode("hello")))
```

```python
print(encode("nikhita"))
print(decode(encode("Nikhita")))
```

```python
words=sorted(list(set(text.lower().split())))
print(len(words))
words.insert(0, "<UNK>")
```

```python
print(words)
```

```python
print("bullshit" in words)
print("good" in words)
```

```python
wtoi={w:i for i,w in enumerate(words)}
itow={i:w for i,w in enumerate(words)}


encode = lambda s: [wtoi.get(w, wtoi["<UNK>"]) for w in s.lower().replace(",", "").replace(".", "").split()]
decode = lambda l: [itow[i] for i in l]
```

```python
print(encode("this is a good project"))
print(decode(encode("this is not good")))
```

```python
print("--- Test 1: Full Paragraph Verification ---")
test_p1 = "This is a good project"
encoded_p1 = encode(test_p1)
decoded_p1 = decode(encoded_p1)
decoded_str_p1 = " ".join(decoded_p1)

print(f"Original: {test_p1}")
print(f"Encoded : {encoded_p1}")
print(f"Decoded : {decoded_str_p1}")
# Verification check (ignoring case for comparison)
print(f"Matches Perfectly?: {test_p1.lower() == decoded_str_p1}\n")


# --- TEST 2: Intentionally breaking it with unknown words ---
print("--- Test 2: Intentionally Breaking with Unknown Words ---")
# 'cat', 'awesome', and 'failed' are not in the vocabulary
test_p2 = "This cat project is awesome and failed"
encoded_p2 = encode(test_p2)
decoded_p2 = decode(encoded_p2)
decoded_str_p2 = " ".join(decoded_p2)

print(f"Original: {test_p2}")
print(f"Encoded : {encoded_p2}")
print(f"Decoded : {decoded_str_p2}")
print("What happened?: The unknown words were safely replaced with '<UNK>' (index 0) instead of crashing!")
```

```python
import importlib
import tiktoken

print("tiktoken version:", importlib.metadata.version("tiktoken"))
# tiktoken version: 0.7.0

tokenizer = tiktoken.get_encoding("gpt2")
text = (
    "Hello, do you like tea?  In the sunlit terraces"
     "of someunknownPlace."
)

integers = tokenizer.encode(text, allowed_special={""})

print(integers)
# [15496, 11, 466, 345, 588, 8887, 30, 220, 50256, 554, 262, 4252, 18250, 8812, 2114, 1659, 617, 34680, 27271, 13]

strings = tokenizer.decode(integers)

print(strings)
# Hello, do you like tea?  In the sunlit terracesof someunknownPlace.
```

```python
from transformers import AutoTokenizer

# 1. Load the tokenizer automatically using the model's Hugging Face shortcut
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-7B")

# 2. Inspect the vocabulary size (which is massive for BPE tokenizers!)
print(f"Vocabulary Size: {tokenizer.vocab_size}")

# 3. Encode a sentence to IDs
text = "Hi I am Nikhita"
encoded_input = tokenizer(text)
print(f"Tokens IDs: {encoded_input['input_ids']}")

# 4. Decode it back to text
decoded_text = tokenizer.decode(encoded_input['input_ids'])
print(f"Decoded: {decoded_text}")
```

```python
inputs=[
    "ಧನ್ಯವಾದಗಳು",
    "Thank you but I don't want to join",
    "solve x^2+4*x=0",
    "class NeuralNetwork(nn.Module): def __init__(self, input_size, hidden_size, output_size) Complete this code"
]


for text in inputs:
    tokens = tokenizer.encode(text)
    print(f"\nText: {text}")
    print(f"Tokens: {tokens}")
    print(f"Count: {len(tokens)}")
    print(f"Decoded pieces: {[tokenizer.decode([t]) for t in tokens]}")
```
